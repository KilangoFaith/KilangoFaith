## **Explore and visualize the data to gain insights**

```python
# visualize the geographical data
housingdf.plot(kind = 'scatter', x = 'longitude', y = 'latitude', figsize= (15,10))
```

```python
# setting the alpha option to 0.1 to visualize the places with high density of data points
housingdf.plot(kind= 'scatter', x= 'longitude', y='latitude', alpha= 0.1, figsize = (15,10))
```

✔ Now high- density areas are clearly seen, Bay Area around Los Angeles, San Diego, Central Valley - around Sacramento and Fresno

```python
# Housing prices view
housingdf.plot (kind = 'scatter', x= 'longitude', y= 'latitude', alpha = 0.4,
 s=housingdf['population']/100, label = 'population', figsize=(15,10),
 c='median_house_value',cmap=plt.get_cmap('jet'), colorbar=True)
plt.legend()
#jet- a predefined color map which ranges from blue to red
```

✔The image tells you that housing prices are very much related to the location and to the pop density.

✔ It is probably useful to use a clustering algorithm to detect the main clusters and add new features that measure the proximity to the cluster centers.

```python
# looking for correlations
# since the dataset is not too large, you can easily compute the std correlation coefficien btwn every pair
# of atrribute
corr_matrix = housingdf.corr()
corr_matrix
```

```python
# how much each attribute correlates with the median house value
corr_matrix['median_house_value'].sort_values(ascending=False)
```

✔The correlation coefficient ranges from -1 to 1. When close to 1, it means that there is a strong positive correlation.

✔Meaning  median house tends to go up when median house values tends to go up. Positive correlation

✔On negative correlation, you can see a small negative correlation beween the latitude and the median house value.

✔Prices close to zero mean that there is no linear correlation

```python
# Another way to check for correlation between attributes is to use Pandas scatter_matrix function
# Scatter_matrix function plots every numerical attribute against every other numerical attribute
attributes = ['median_house_value', 'median_income', 'total_rooms', 'housing_median_age']
scatter_matrix(housingdf[attributes], figsize = (12,8))
```

```python
# The most promising attribute to predict the median house value is the median income, so lets zoom in their correlation
housingdf.plot(kind= 'scatter', x= 'median_income', y= 'median_house_value',alpha =0.1, figsize= (15,10))
```

✔There is strong correlation, hence upward trend and points are not too dispersed

# Data Quality Verification


```python
# missing values - apply pandas info() or isnull() function
housingdf.isnull().sum()
```

```python
# outlier values - extreme values
fig = plt.figure(figsize= (10,7))
plt.boxplot(housingdf['households'])
```

```python
# we can also plot all boxplots for all features. Excluding the non numerical variables
for col in['longitude','latitude','housing_median_age','total_rooms','total_bedrooms',
           'population','households','median_income','median_house_value']:
           print(col)
           plt.boxplot(housingdf[col])
           plt.show()
```

✔ All features except longitude, latitude and housing _median_age have outliers
